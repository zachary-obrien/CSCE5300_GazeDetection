```python
import glob
import json
import pandas
import cv2
```

```python
import pandas as pd

data_folder = 'E://eyetracker/Extracts/'
input_data = None
for filename in glob.glob(data_folder + '*.csv'):
    input_data = pd.read_csv(filename, dtype={'Sample': object})
```

```python
#input_data.reset_index()
tmp_storage = {}
with open("temp_sex.json") as json_file:
    tmp_storage = json.load(json_file)
input_data.sort_values(by="Sample", ascending=True, inplace=True)
for index, row in input_data.iterrows():
    sample_id = row['Sample']
    if sample_id in tmp_storage:
        continue
    sample_image = data_folder + sample_id + "/frames/00001.jpg"
    img = cv2.imread(sample_image)
    cv2.imshow("test", img)
    k = cv2.waitKey(0)
    print(sample_id, "is", k)
    if k%256 == 48:
        tmp_storage[sample_id] = 0
    elif k%256 == 49:
        tmp_storage[sample_id] = 1
    elif k%256 == 27:
        # ESC pressed
        print("Escape hit, closing...")
        break
    if (len(tmp_storage) % 10) == 0:
        with open("temp_sex.json", "w") as outfile:
            json.dump(tmp_storage, outfile)
with open("temp_sex.json", "w") as outfile:
            json.dump(tmp_storage, outfile)
print("Done")
cv2.destroyAllWindows()
```

```python
print(len(tmp_storage))
df = pd.DataFrame(columns=['Sample', 'Sex'])
for key in tmp_storage.keys():
    df = df.append({'Sample':key, 'Sex':tmp_storage[key]}, ignore_index=True)
```

```python
print(df)
for filename in glob.glob(data_folder + '*.csv'):
    input_data = pd.read_csv(filename, dtype={'Sample': object})
    merged_left = pd.merge(left=input_data, right=df, how='left', left_on='Sample', right_on='Sample')
    print(merged_left)
    new_filename = filename.replace(".csv", "with_sex.csv")
    merged_left.to_csv(new_filename,index=False)
```
